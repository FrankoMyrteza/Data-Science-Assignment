```python
import pandas
import pylab as pl
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
variables = pandas.read_csv("C:/Users/Fmyrt/Desktop/netflow_day-02.csv",sep=",", index_col=False , nrows=1000)

Y = variables[['SrcPackets']]

X = variables[['SrcBytes']]

W = variables[['DstBytes']]

Z = variables[['DstPackets']]

SrcBytes = range(1, 20)

kmeans = [KMeans(n_clusters=i) for i in SrcBytes]

kmeans

SrcPackets = [kmeans[i].fit(Y).score(Y) for i in range(len(kmeans))]

SrcPackets

pl.plot(SrcBytes,SrcPackets)

pl.xlabel('k')

pl.ylabel('Distortion')

pl.title('Elbow Curve')

pl.show()

pca = PCA(n_components=1).fit(Y)

pca_d = pca.transform(Y)

pca_c = pca.transform(X)

kmeans=KMeans(n_clusters=3)

kmeansoutput=kmeans.fit(Y)

kmeansoutput

pl.figure('3 Cluster K-Means')

pl.scatter(pca_c[:, 0], pca_d[:, 0], c=kmeansoutput.labels_)

pl.xlabel('Source Packets')

pl.ylabel('Source Bytes')

pl.title('3 Cluster K-Means')

pl.show()

pca = PCA(n_components=1).fit(W)

pca_d = pca.transform(W)

pca_c = pca.transform(Z)

kmeans=KMeans(n_clusters=3)

kmeansoutput=kmeans.fit(W)

kmeansoutput

pl.figure('3 Cluster K-Means')

pl.scatter(pca_c[:, 0], pca_d[:, 0], c=kmeansoutput.labels_)

pl.xlabel('Destination Packets')

pl.ylabel('Destination Bytes')

pl.title('3 Cluster K-Means')

pl.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings; warnings.simplefilter('ignore')
%matplotlib inline
```

```python
x = credit[['SrcPackets','DstPackets','SrcBytes','DstBytes']]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_normalized = pd.DataFrame(scaler.fit_transform(x))
```

```python
x_normalized.columns = ['SrcPackets','DstPackets','SrcBytes','DstBytes']
```

```python
from sklearn.ensemble import IsolationForest
iforest = IsolationForest()
iforest.fit(x_normalized)
```

```python
Anomaly_Class = iforest.predict(x_normalized)
Anomaly_Score = iforest.decision_function(x_normalized)
```

```python
x['Anomaly_Class'] = Anomaly_Class
x['Anomaly_Score'] = Anomaly_Score
```

```python
x.groupby('Anomaly_Class')[['Anomaly_Score']].count()
```

```python
x['Anomaly_Score'].hist(by=x['Anomaly_Class'])
```

```python
x['Anomaly_Score'].describe()
```

```python
from sklearn.ensemble import IsolationForest
iforest = IsolationForest(bootstrap=True)
iforest.fit(x_normalized)
```
